## 모델, 토크나이저 불러오기

```python
from transformers import AutoTokenizer, AutoModelForCausalLM

model_name = "openai/gpt-oss-20b"

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(
    model_name,
    torch_dtype="auto",
    device_map="auto"
)
```

```python
tokenizer
```

```python
model
```

```python
model.loss_function
```

## input 토큰화

```python
messages = [
    {"role": "user", "content": "Explain what MXFP4 quantization is."},
]

inputs = tokenizer.apply_chat_template(
    messages,
    add_generation_prompt=True,
    return_tensors="pt",
    return_dict=True,
).to(model.device)
```

```python
inputs
```

```python
len(inputs.input_ids[0]), len(inputs.attention_mask[0])
```

## 일반 generate

output은 키/값들이 openai delimiter로 구분되어 있다.

openai-harmony 라이브러리로 output 파싱을 해줄 수도 있다.
- 참고: https://cookbook.openai.com/articles/gpt-oss/run-transformers#chat-template-and-tool-calling


```python
outputs = model.generate(
    **inputs,
    max_new_tokens=200,
    temperature=0.7
)

tokenizer.decode(outputs[0])
```

## Stream 모드 사용하기

> 참고:
- 블로그: https://hypro2.github.io/transformers-generator/
- Transformers.Streamer: https://hypro2.github.io/transformers-generator/



```python
from threading import Thread
from transformers import TextIteratorStreamer

streamer = TextIteratorStreamer(tokenizer, timeout=10., skip_prompt=True, skip_special_tokens=True)
generate_kwargs = dict(
    inputs,
    streamer=streamer,
    max_new_tokens=200,
    temperature=0.7,
)

# 스레드를 따로 써야한다
t = Thread(target=model.generate, kwargs=generate_kwargs)
t.start()

model_output = ""
for new_text in streamer:
    model_output += new_text
    print(new_text, end="", flush=True)
```
